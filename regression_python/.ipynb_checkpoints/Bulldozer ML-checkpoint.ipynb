```python
#create a root msle fct
def rmsle(y_true,y_pred):
    return np.sqrt(msle(y_true,y_pred))
```

```python
#evaluating model fct
from sklearn.metrics import mean_absolute_error as mae
```

```python
def eval0(model):
    train_pred = model.predict(X_train)
    valid_pred = model.predict(X_valid)
    dict_met = {"MAE metric train score":mae(y_train,train_pred),
               "MAE metric valid score":mae(y_valid,valid_pred),
               "RMSLE metric train score":rmsle(y_train,train_pred),
               "RMSLE metric valid score":rmsle(y_valid,valid_pred),
               "R2 metric train score":r2_score(y_train,train_pred),
               "R2 metric valid score":r2_score(y_valid,valid_pred)}
    return dict_met
```

```python
eval0(modelf)
```

```python
def eval1(model):
    train_pred = model.predict(X_train)
    valid_pred = model.predict(X_valid)
    dict_met = {"MAE metric train score":mae(y_train,train_pred),
               "MAE metric valid score":mae(y_valid,valid_pred),
               "R2 metric train score":r2_score(y_train,train_pred),
               "R2 metric valid score":r2_score(y_valid,valid_pred)}
    return dict_met
```

## Make prediction on the test data

### preprocessing test data

```python
def preprocess_data(df):
    #datetime engineering to give weight to year,month and day of when the sales happened
    df["saleyear"] = df["saledate"].dt.year
    df["salemonth"] = df["saledate"].dt.month
    df["saleday"] = df["saledate"].dt.day
    df["saledayofyear"] = df["saledate"].dt.dayofyear
    df["saledayofweek"] = df["saledate"].dt.dayofweek
    df.drop("saledate",axis=1,inplace=True)
    
    #turning string values into categorical values!!!
    for key,value in df.items():
        if pd.api.types.is_numeric_dtype(value):
            if pd.isnull(value).sum():
                df[key+"_is_missing"] = pd.isnull(value)
                df[key] = value.fillna(value.median()) 
        if not pd.api.types.is_numeric_dtype(value):    
            df[key+"_is_missing"] = pd.isnull(value)
            df[key] = pd.Categorical(value).codes + 1 
    return df
    
    
    
```

### Feature importance

```python
#let's view feature importance in our trained model
ideal_model.feature_importances_
```

```python
#let us fisualize the feature importance using a helper function
def plot_feat_imp(column,feat_imp,n=20):
    #first we define a dataframe containing feature name and their importance
    df = (pd.DataFrame({"features":column,
                       "feature importance":feat_imp})
                     .sort_values("feature importance",ascending=False)
                     .reset_index(drop=True))
    fig,ax = plt.subplots(figsize=[10,8])
    ax.barh(df["features"][:n],df["feature importance"][:n])
    ax.set_ylabel("features")
    ax.set_xlabel("feature importances")
    ax.invert_yaxis() #if we had ascending we will get descending or vice versa
```

```python
plot_feat_imp(X_train.columns,ideal_model.feature_importances_)
```
